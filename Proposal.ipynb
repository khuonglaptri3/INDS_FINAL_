# 1. Giới thiệu bài toán và dữ liệu

## 1.1. Động lực nghiên cứu
<div style="text-align: justify;">

- Thu nhập cá nhân là một khía cạnh quan trọng trong nền kinh tế, phản ánh cơ hội nghề nghiệp, trình độ giáo dục và cấu trúc thị trường lao động. Việc hiểu các yếu tố liên quan đến thu nhập có ý nghĩa lớn trong nghiên cứu xã hội học và khoa học dữ liệu.

- Nghiên cứu của nhóm tập trung vào việc khám phá các yếu tố nhân khẩu học và nghề nghiệp tác động đến vị thế kinh tế của cá nhân thông qua bộ dữ liệu Adult Census (1994). Thay vì chỉ coi đây là một bài toán phân loại nhị phân đơn thuần, cách tiếp cận này mang tính thực tiễn hơn, vì bộ dữ liệu chưa phản ánh toàn bộ các yếu tố ảnh hưởng tới thu nhập thực sự của một cá nhân.

- Bên cạnh đó, bộ dữ liệu là môi trường phù hợp để nghiên cứu các kỹ thuật tiền xử lý phức tạp như xử lý dữ liệu thiếu và mất cân bằng lớp, đồng thời là môi trường tốt để nhóm thực hành các kỹ thuật EDA. Cuối cùng, nghiên cứu hướng tới phát triển các mô hình có khả năng giải thích (XAI) và đảm bảo tính công bằng, tránh khuếch đại các định kiến lịch sử trong dữ liệu.

</div>

<div style="text-align: justify;">

- Mô hình có phản ánh các định kiến lịch sử liên quan đến giới tính hoặc sắc tộc hay không? Làm thế nào để diễn giải các quyết định dự báo của mô hình thông qua các phương pháp như SHAP hoặc LIME?

- Khi kiểm soát các yếu tố như số giờ làm việc và trình độ học vấn, các đặc điểm như giới tính và sắc tộc ảnh hưởng như thế nào đến khả năng đạt được mức thu nhập cao?

# 2. Kế hoạch phân tích dữ liệu

## 2.1. Input và Output của bài toán

- **Biến mục tiêu (target) Y** là biến nhị phân *income*:
  - \( Y = 1 \): nếu thu nhập > 50.000 USD  
  - \( Y = 0 \): nếu thu nhập ≤ 50.000 USD  

- **Biến giải thích (features) X** bao gồm 14 biến (cả dạng **categorical** và **numerical**).

- Nhóm sẽ thực hiện phân tích khám phá dữ liệu (EDA) nhằm:
  - Loại bỏ các đặc trưng không mang ý nghĩa dự báo đối với từng cá nhân (giảm nhiễu)  
  - Xử lý các biến có tương quan cao (tránh trùng lặp thông tin)  
  - Khai thác các insight quan trọng từ dữ liệu để hỗ trợ xây dựng mô hình  

## 2.2. Các phương pháp đánh giá

- Kiểm tra mức độ **mất cân bằng dữ liệu (class imbalance)**.


- Bên cạnh **Accuracy**, nhóm tập trung vào các thước đo phù hợp hơn:
  - **F1-Score** (cân bằng giữa Precision và Recall)  
  - **AUC-ROC**  
  - **MCC (Matthews Correlation Coefficient)**  

- Sử dụng **Stratified K-Fold Cross Validation** để đảm bảo việc đánh giá mô hình ổn định và đáng tin cậy.

- Áp dụng các phương pháp **giải thích mô hình (Explainable AI - XAI)** nhằm phân tích hành vi của mô hình ở cả mức độ toàn cục và cục bộ:

  - **SHAP (SHapley Additive exPlanations)**:
    - Sử dụng *TreeExplainer* (phù hợp với mô hình LightGBM)
    - Phân tích:
      - *Summary plot (bar)*: xếp hạng mức độ quan trọng của các đặc trưng
      - *Summary plot (dot)*: thể hiện chiều tác động (tăng/giảm) của từng đặc trưng
      - *Dependence plot*: phân tích tương tác giữa các biến (ví dụ: *education-num* và *age*)
      - *Waterfall plot*: giải thích dự đoán cho từng cá thể cụ thể (đặc biệt là các trường hợp dự đoán sai)
      - *Bias analysis*: đánh giá ảnh hưởng của các biến nhạy cảm (*sex*, *race*)

  - **LIME (Local Interpretable Model-agnostic Explanations)**:
    - Giải thích cục bộ cho từng quan sát bằng cách xấp xỉ mô hình phức tạp bằng mô hình tuyến tính trong vùng lân cận
    - Tập trung phân tích các trường hợp dự đoán sai để hiểu nguyên nhân sai lệch

  - **DiCE (Diverse Counterfactual Explanations)**:
    - Sinh các *counterfactual explanations* nhằm trả lời câu hỏi:
      > “Cần thay đổi những đặc trưng nào để đảo ngược kết quả dự đoán?”
    - Hỗ trợ phân tích tính hợp lý và khả năng can thiệp của mô hình


- Đánh giá **tính công bằng (fairness)** của mô hình thông qua các chỉ số:

  - **Demographic Parity Difference (DPD)**:
    - Đo lường chênh lệch tỷ lệ dự đoán dương giữa các nhóm

  - **Demographic Parity Ratio (DPR)**:
    - Tỷ lệ dự đoán giữa các nhóm (giá trị lý tưởng gần 1.0)

  - **Equalized Odds Difference (EOD)**:
    - Đo lường sự khác biệt về *True Positive Rate* và *False Positive Rate* giữa các nhóm

- Các chỉ số fairness được đánh giá trên các thuộc tính nhạy cảm như:
  - **Giới tính (sex)**
  - **Sắc tộc (race)**

- Sử dụng các kỹ thuật trực quan hóa:
  - Ma trận tương quan (**correlation matrix**)  
  - Biểu đồ nhiệt (**heatmap**)  
  → để phân tích mối quan hệ giữa các biến trong tập dữ liệu.

## 2.3. Các mô hình học máy sử dụng

Để giải quyết bài toán phân loại thu nhập, nhóm sử dụng và so sánh các mô hình học máy sau:

- **Logistic Regression**:
  - Mô hình tuyến tính cơ bản, dễ diễn giải
  - Được sử dụng làm baseline để so sánh với các mô hình phức tạp hơn

- **Random Forest Classifier**:
  - Mô hình ensemble dựa trên nhiều cây quyết định
  - Giảm overfitting và cải thiện độ ổn định so với single decision tree

- **LightGBM (Light Gradient Boosting Machine)**:
  - Mô hình boosting hiệu quả, tối ưu cho dữ liệu lớn và nhiều đặc trưng
  - Tốc độ huấn luyện nhanh và khả năng xử lý dữ liệu dạng bảng tốt

- **XGBoost (Extreme Gradient Boosting)**:
  - Mô hình boosting mạnh mẽ với khả năng regularization tốt
  - Thường đạt hiệu suất cao trong các bài toán phân loại

Các mô hình sẽ được đánh giá và so sánh dựa trên các tiêu chí đã đề cập (F1-score, AUC-ROC), đồng thời phân tích thêm về khả năng diễn giải và tính công bằng.

# 3. Những kết quả mong đợi và thách thức


## 3.1. Những kết quả mong đợi

- Xác định được các yếu tố then chốt trong bộ dữ liệu có ảnh hưởng mạnh và mang tính quyết định đến xác suất đạt mức thu nhập cao.

- Thông qua các phương pháp giải thích mô hình như **SHAP** và **LIME**, nghiên cứu kỳ vọng trực quan hóa được cơ chế ra quyết định của mô hình. Ví dụ: xác định ngưỡng số năm học vấn có thể giúp một cá nhân chuyển từ nhóm thu nhập thấp sang thu nhập cao.

- Cung cấp các bằng chứng định lượng về khoảng cách thu nhập giữa các nhóm:
  - Giữa **nam và nữ**
  - Giữa các **sắc tộc khác nhau**  
  Sau khi đã kiểm soát các yếu tố như nghề nghiệp và số giờ làm việc.

## 3.2. Các thách thức và hạn chế

- **Tính lỗi thời của dữ liệu**:  
  Dữ liệu được thu thập từ năm 1994, do đó các quy luật về thu nhập có thể đã thay đổi đáng kể do các yếu tố như lạm phát, sự phát triển của các ngành công nghệ mới và thay đổi trong cấu trúc kinh tế – xã hội.

- **Đơn giản hóa bài toán phân loại**:  
  Việc chia thu nhập thành hai lớp (≤50K và >50K) có thể làm mất đi thông tin chi tiết về phân phối thu nhập, từ đó hạn chế khả năng phản ánh đầy đủ mức độ bất bình đẳng kinh tế.

- **Thiếu biến quan trọng**:  
  Bộ dữ liệu không bao gồm các yếu tố có ảnh hưởng lớn đến thu nhập thực tế như:
  - Kỹ năng mềm  
  - Vị trí địa lý  
  - Tài sản thừa kế  
  - Mạng lưới quan hệ xã hội  

- **Vấn đề dữ liệu (missing & imbalance)**:  
  Sự tồn tại của giá trị thiếu và mất cân bằng lớp đòi hỏi áp dụng các kỹ thuật xử lý như:
  - Imputation  
  - SMOTE  
  - Class weighting  
  nhằm giảm thiểu sai lệch trong quá trình huấn luyện mô hình.

- **Định kiến trong dữ liệu (historical bias)**:  
  Dữ liệu có thể phản ánh các định kiến xã hội tồn tại trong thập niên 1990 tại Hoa Kỳ. Nếu không được kiểm soát, mô hình có thể học và khuếch đại các sai lệch này, gây khó khăn cho việc đảm bảo tính công bằng.
