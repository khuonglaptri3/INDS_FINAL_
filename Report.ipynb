### Quy trình xây dựng mô hình

Hình trên thể hiện quy trình tổng thể của hệ thống học máy được đề xuất trong nghiên cứu, bao gồm các bước tuần tự từ xử lý dữ liệu đầu vào đến phân tích giải thích mô hình. Cụ thể, dữ liệu ban đầu (raw data) được đưa qua giai đoạn tiền xử lý nhằm làm sạch, xử lý giá trị thiếu, mã hóa các biến phân loại và chuẩn hóa các biến số. Bước này đóng vai trò quan trọng trong việc đảm bảo dữ liệu đầu vào phù hợp với các thuật toán học máy và hạn chế hiện tượng rò rỉ dữ liệu (data leakage).

Sau đó, dữ liệu đã được xử lý được sử dụng để huấn luyện các mô hình phân loại, bao gồm cả mô hình tuyến tính và các mô hình phi tuyến thuộc nhóm ensemble. Quá trình huấn luyện được thiết kế theo pipeline nhất quán nhằm đảm bảo tính công bằng trong so sánh và khả năng tái lập kết quả. 

Hiệu suất của các mô hình được đánh giá thông qua các độ đo chính là F1-score và AUC. Việc lựa chọn hai chỉ số này nhằm đảm bảo mô hình không chỉ có khả năng phân biệt tốt giữa các lớp mà còn duy trì sự cân bằng giữa Precision và Recall trong bối cảnh dữ liệu mất cân bằng.

Cuối cùng, để vượt qua hạn chế của các mô hình “hộp đen”, nghiên cứu tích hợp các phương pháp giải thích mô hình (Explainable AI), bao gồm phân tích mức độ quan trọng của đặc trưng và các kỹ thuật như SHAP, LIME và DiCE. Các phương pháp này cho phép phân tích cả ở mức toàn cục và cục bộ, từ đó cung cấp cái nhìn sâu sắc về cơ chế ra quyết định của mô hình, đồng thời hỗ trợ các phân tích nâng cao như đánh giá bias và fairness.

Tổng thể, quy trình này được xây dựng theo hướng toàn diện, đảm bảo sự kết hợp giữa hiệu suất dự đoán, khả năng tổng quát hóa và tính minh bạch trong giải thích mô hình.

<div align="center">
  <img src="image/Preprocessing_Fearture.png" width="1400" height="800" />

- Quy trình bắt đầu từ dữ liệu thô, thực hiện xử lý giá trị thiếu bằng mode và feature engineering, sau đó chia train/test (80/20), áp dụng SMOTE để cân bằng dữ liệu, huấn luyện mô hình và đánh giá bằng các chỉ số như F1-score và Accuracy.


<div align="center">

Pipeline feature engineering được thiết kế theo nguyên tắc **không rò rỉ dữ liệu (leak-free)**, trong đó toàn bộ các phép biến đổi chỉ được *fit trên tập train* và sau đó áp dụng lên tập test. Cách tiếp cận này đảm bảo tính khách quan trong đánh giá mô hình và duy trì khả năng tổng quát hóa. Quy trình bao gồm bốn thành phần chính: biến đổi biến số, mã hóa biến phân loại, tạo đặc trưng tương tác và kiểm soát data leakage.

Các biến số liên tục có phân phối không đồng nhất, đặc biệt là các biến tài chính như *capital-gain* và *capital-loss* có độ lệch phải cao. Để xử lý vấn đề này, nhóm áp dụng **biến đổi Yeo-Johnson**, cho phép đưa phân phối về gần chuẩn mà không yêu cầu dữ liệu phải dương. Các tham số ước lượng tương ứng là λ = -1.37 cho *capital-gain* và λ = -2.85 cho *capital-loss*, giúp ổn định phương sai và tăng độ nhạy của mô hình đối với các giá trị nhỏ.

Để giảm ảnh hưởng của ngoại lai (outliers), các biến *age*, *hours-per-week* và *education-num* được chuẩn hóa bằng **Robust Scaling**, được định nghĩa như sau:

$$
X_{scaled} = \frac{X - \mathrm{median}(X)}{\mathrm{IQR}(X)}
$$

Trong đó **IQR (Interquartile Range)** là khoảng tứ phân vị (Q3 - Q1). Phương pháp này giúp mô hình ổn định trước outliers và vẫn giữ được cấu trúc tương đối của dữ liệu.

Đối với các biến phân loại, chiến lược mã hóa được thiết kế linh hoạt theo đặc điểm của từng biến. Các biến có số lượng giá trị lớn như *occupation* được gom nhóm thành các cụm có ý nghĩa (manual, high-skill, office-tech, specialised), sau đó mã hóa bằng **CatBoost Encoding** nhằm giảm thiểu hiện tượng rò rỉ dữ liệu. Biến *workclass* được xử lý bằng **Leave-One-Out (LOO) Encoding**, cung cấp ước lượng trung bình target đã được điều chỉnh để hạn chế overfitting.

Biến *native-country* được chuyển đổi sang dạng thứ bậc thông qua biến *country_income_group*, dựa trên phân loại thu nhập quốc gia, giúp bảo toàn thông tin thứ tự. Các biến phân loại có số lượng nhỏ như *relationship* và *race* được mã hóa bằng **one-hot encoding**, trong khi các biến nhị phân như *sex* và *marital-status* được chuyển thành các biến chỉ báo (binary indicators) dễ diễn giải.

Để nắm bắt các quan hệ phi tuyến và tương tác giữa các biến, pipeline tạo thêm **18 đặc trưng mới**. Các đặc trưng kinh tế (econometric features) bao gồm *human_capital* (age × education), *household_labour* (hours × marital status) và *net_capital* (capital gain trừ capital loss), giúp tóm tắt các yếu tố kinh tế quan trọng. Bên cạnh đó, các **đặc trưng tương tác phục vụ phân tích công bằng (fairness interactions)** được xây dựng bằng cách kết hợp biến *race* với các biến như *education*, *hours-per-week* và *net_capital*, nhằm hỗ trợ phân tích sự khác biệt giữa các nhóm nhân khẩu học.

Không gian đặc trưng cuối cùng bao gồm khoảng 40 biến, kết hợp giữa các biến số đã biến đổi, biến phân loại đã mã hóa và các đặc trưng tương tác. Đáng chú ý, phần lớn các đặc trưng mới được tạo ra từ **các biến tương tác và các biến phục vụ phân tích fairness**, thay vì chỉ đơn thuần từ các kỹ thuật mã hóa.

Để đảm bảo không xảy ra rò rỉ dữ liệu, toàn bộ các bước tiền xử lý—bao gồm các phép biến đổi và encoder—đều được huấn luyện trên tập train (39,073 mẫu) và chỉ áp dụng lên tập test (9,769 mẫu). Điều này đảm bảo rằng kết quả đánh giá phản ánh đúng năng lực dự đoán của mô hình trên dữ liệu chưa từng quan sát.

- Về xử lý mất cân bằng lớp bằng smote

- Quá trình sử dụng  smote đã  thay đổi trong phân phối lớp của tập dữ liệu trước và sau khi áp dụng kỹ thuật SMOTE. Cụ thể, trong tập huấn luyện ban đầu, dữ liệu bị mất cân bằng đáng kể khi lớp thu nhập ≤50K chiếm tỷ lệ 76.1%, trong khi lớp >50K chỉ chiếm 23.9%. - Việc tái cân bằng này đóng vai trò quan trọng trong quá trình huấn luyện mô hình, giúp giảm sự thiên lệch về lớp đa số và buộc mô hình học được các đặc trưng đại diện tốt hơn cho lớp thu nhập cao. Nhờ đó, khả năng phát hiện lớp thiểu số được cải thiện đáng kể, đặc biệt thể hiện qua sự gia tăng của chỉ số Recall và F1-score, đồng thời tạo nền tảng đáng tin cậy cho các phân tích nâng cao như giải thích mô hình và đánh giá công bằng.

### 2.2 Kiến trúc mô hình và quá trình huấn luyện

<div align="center">
  <img src="image/config.png" />
  <p><i>Figure 6: Model Configuration </i></p>
</div>


Trong nghiên cứu này, nhóm không thực hiện fine-tuning toàn diện để tối ưu siêu tham số (hyperparameters) ở giai đoạn đầu, mà thay vào đó sử dụng các cấu hình được tham khảo trực tiếp từ các nghiên cứu trước. Cách tiếp cận này nhằm đảm bảo tính tái lập (reproducibility) và cho phép so sánh kết quả một cách khách quan với các công trình liên quan. Do đó, các cấu hình được xem như một baseline hợp lý và có cơ sở học thuật, trước khi tiến hành tinh chỉnh sâu hơn nếu cần thiết.

- Cụ thể, các mô hình đều được thiết lập với 300 estimators và learning rate = 0.05, nhằm cân bằng giữa tốc độ hội tụ và độ ổn định trong quá trình huấn luyện. Độ sâu của cây (max depth) được giới hạn (chủ yếu ở mức 8) nhằm kiểm soát hiện tượng overfitting nhưng vẫn đảm bảo khả năng học các quan hệ phi tuyến.

- Đối với các mô hình boosting như XGBoost và LightGBM, regularization L1 (0.1) và L2 (1.0) được áp dụng để kiểm soát độ phức tạp mô hình. Đồng thời, các tham số như subsample = 0.8 và colsample_bytree = 0.8 được sử dụng để đưa vào tính ngẫu nhiên, giúp giảm overfitting và tăng khả năng tổng quát hóa.

- LightGBM tận dụng các cơ chế tối ưu như Gradient-based One-Side Sampling và Exclusive Feature Bundling để cải thiện hiệu suất tính toán, trong khi vẫn giữ cấu hình regularization tương tự nhằm đảm bảo tính nhất quán trong so sánh.

- Random Forest được thiết lập với độ sâu lớn hơn (max depth = 12) nhằm tận dụng đặc tính ensemble của nhiều cây độc lập. Các tham số như min_samples_split = 10 và min_samples_leaf = 5 được sử dụng để kiểm soát sự phát triển của cây, trong khi max_features = sqrt giúp tăng tính đa dạng giữa các cây.

- Đối với Logistic Regression, mô hình được sử dụng với penalty L2, hệ số điều chuẩn C = 1, solver liblinear và số vòng lặp tối đa là 2000, đóng vai trò như một baseline tuyến tính đơn giản nhưng ổn định.

- Nhìn chung, việc sử dụng cấu hình từ các bài báo giúp đảm bảo nền tảng thực nghiệm vững chắc. Trong các bước tiếp theo, nếu hiệu suất mô hình chưa đạt yêu cầu, nhóm có thể tiến hành fine-tuning có kiểm soát nhằm tối ưu thêm hiệu năng mà vẫn giữ được tính so sánh và minh bạch của nghiên cứu.

### 2.3 Explainable AI và  Fairness Analysis

<div align="center">
  <img src="image/Fow_EXPLAIN_AI.png" />
  <p><i>Figure 7: Sơ đồ mô tả những phương pháp mà  nhóm thực hiện đối với giải thích mô hinh và các phân tích công bằng. </i></p>

Để đảm bảo khả năng diễn giải và đánh giá tính công bằng của mô hình, nghiên cứu áp dụng một pipeline Explainable AI kết hợp giữa các phương pháp giải thích toàn cục và cục bộ, bao gồm SHAP, Fairlearn, LIME và DiCE. Cách tiếp cận này cho phép phân tích không chỉ cách mô hình hoạt động tổng thể mà còn lý do đằng sau từng dự đoán cụ thể.

- Đối với giải thích toàn cục, phương pháp SHAP (SHapley Additive exPlanations) được triển khai thông qua TreeExplainer, phù hợp với mô hình LightGBM dạng cây. SHAP được sử dụng để xây dựng các biểu đồ feature importance nhằm xếp hạng mức độ ảnh hưởng trung bình của các biến, đồng thời sử dụng summary plot để thể hiện phân phối giá trị SHAP và chiều tác động của từng đặc trưng lên dự đoán. Bên cạnh đó, dependence plot được khai thác để phân tích tương tác giữa các biến quan trọng như education-num và age. Đặc biệt, nghiên cứu thực hiện SHAP-based bias analysis bằng cách kiểm tra ảnh hưởng của các biến nhạy cảm (sex_binary, race_White) trong điều kiện đã kiểm soát các yếu tố như trình độ học vấn và số giờ làm việc. Ngoài ra, waterfall plot được sử dụng để giải thích chi tiết các trường hợp dự đoán sai ở mức cá thể.

- Song song với đó, tính công bằng của mô hình được đánh giá thông qua thư viện Fairlearn với các chỉ số tiêu chuẩn. Cụ thể, Demographic Parity Difference (DPD) đo lường sự chênh lệch xác suất dự đoán dương giữa các nhóm, Demographic Parity Ratio (DPR) phản ánh tỷ lệ bất lợi tương đối của nhóm yếu thế, và Equalized Odds Difference (EOD) đánh giá sự khác biệt về tỷ lệ dự đoán đúng (TPR) và sai (FPR) giữa các nhóm nhân khẩu học. Các chỉ số này cung cấp góc nhìn định lượng về mức độ thiên lệch mà mô hình có thể kế thừa từ dữ liệu.

- Ở  mức giải thích cục bộ, phương pháp LIME được áp dụng thông qua LimeTabularExplainer để phân tích một instance cụ thể, đặc biệt là các trường hợp dự đoán sai. LIME cho phép xác định các đặc trưng có ảnh hưởng lớn nhất tại điểm dữ liệu đó, từ đó cung cấp một mô hình xấp xỉ tuyến tính dễ hiểu trong lân cận của instance.

- Cuối cùng, phương pháp DiCE (Diverse Counterfactual Explanations) được sử dụng để sinh các counterfactual instances, trả lời câu hỏi “cần thay đổi những đặc trưng nào để đảo ngược dự đoán của mô hình”. Thông qua thuật toán di truyền, DiCE tạo ra nhiều kịch bản thay thế với desired_class đối lập, giúp làm rõ ranh giới quyết định của mô hình và cung cấp thông tin hữu ích cho phân tích hành vi mô hình.

- Tổng thể, pipeline này kết hợp hiệu quả giữa giải thích toàn cục (SHAP, fairness metrics) và giải thích cục bộ (SHAP waterfall, LIME, DiCE), từ đó xây dựng một hệ thống không chỉ có hiệu năng cao mà còn minh bạch và đáng tin cậy trong bối cảnh các bài toán liên quan đến dữ liệu nhạy cảm.

### 3.2 Explainable AI và  Fairness Analysis

- SHAP cho thấy mức độ ảnh hưởng của các biến được phân cấp rất rõ ràng. Từ biểu đồ feature importance, married_flag (~1.14) có tác động trung bình lớn nhất, tiếp theo là age (~0.80). Điều này cho thấy các yếu tố phản ánh giai đoạn cuộc sống và vị thế xã hội có sức nặng cao hơn đáng kể so với số năm học; tác động của tuổi gần gấp đôi học vấn.

- Biểu đồ beeswarm làm rõ tác động ở mức cá nhân. Với age, các điểm màu xanh (tuổi thấp) tạo đuôi SHAP âm rất sâu, có thể xuống khoảng −4, làm giảm mạnh xác suất thu nhập cao. Ngược lại, các điểm màu hồng (tuổi cao hơn) tập trung vùng SHAP dương khoảng +1 đến +2. Biến married_flag hầu như luôn nằm phía dương, còn capital-gain xuất hiện một số điểm cực trị có thể đạt ~+5, cho thấy tích lũy tài chính có thể làm thay đổi dự đoán rất mạnh trong một số trường hợp.

- Biểu đồ dependence của age cho thấy dạng inverted-U rõ rệt: giá trị SHAP tăng mạnh từ vùng chuẩn hóa khoảng −1 → 0.5, đạt đỉnh khoảng 0.5–1.0, rồi giảm nhẹ sau ~1.5. Khi kết hợp màu theo hours-per-week, nhóm trung niên làm nhiều giờ (màu hồng) tập trung ở vùng SHAP dương cao nhất.

Tổng hợp lại, age nổi bật vì vừa có mức ảnh hưởng lớn (~0.80), vừa tạo biên độ tác động rộng (−4 → +2) và thể hiện ngưỡng phi tuyến rõ ràng. Điều này cho thấy mô hình đang phản ánh cấu trúc phân phối thu nhập trong dữ liệu xã hội–kinh tế, nơi kinh nghiệm, tích lũy và giai đoạn sự nghiệp đóng vai trò then chốt.

Mô hình không chỉ phản ánh bất bình đẳng lịch sử trong dữ liệu mà còn có nguy cơ duy trì chúng thông qua cấu trúc quan hệ giữa các biến. Bias vì vậy không nằm ở một biến riêng lẻ, mà nằm ở toàn bộ hệ thống dữ liệu mà mô hình đã học được, và đây ` là điều mà nhóm bất ngờ nhất ` sau những kết quả phân tích nhận được.

Các yếu tố mang tính xã hội–nhân khẩu học đóng vai trò lực đẩy mạnh, nổi bật là married_flag (+0.94) và age (+0.54), trong khi các yếu tố năng lực và tích lũy lại tạo lực cản đáng kể như workclass (−0.73), education-num (−0.46) và capital-gain = 0 (−0.36). Sự xuất hiện của biến tương tác như capital_x_Amer-Indian-Eskimo với hệ số âm lớn cho thấy mô hình so sánh cá nhân với các nhóm tham chiếu trong không gian dữ liệu; quyết định không chỉ dựa trên năng lực cá nhân mà còn phụ thuộc vào đặc điểm nhân khẩu học liên quan đến cơ hội tài chính.

- Với  DiCE làm rõ mức độ nhạy của ranh giới quyết định theo tuổi. Instance gốc có age chuẩn hóa = −0.35 bị dự đoán thu nhập thấp, nhưng chỉ cần tăng nhẹ tuổi lên −0.30, −0.10 hoặc 0.0 là dự đoán chuyển sang thu nhập cao, trong khi các đặc trưng năng lực khác giữ nguyên. Điều này cho thấy tồn tại một “ngưỡng quyết định” rất nhạy theo tuổi, nơi thay đổi nhỏ cũng đủ làm đảo chiều kết quả. Yếu tố cuối cùng là  SHAP Waterfall mô tả chi tiết cơ chế cộng dồn đóng góp đặc trưng: từ giá trị cơ sở E[f(X)] = 0.144 đến dự đoán cuối f(x) = 0.191. Lực đẩy lớn nhất đến từ married_flag (+0.94), age (+0.54) và relationship_Husband (+0.12); trong khi lực cản mạnh nhất là workclass (−0.73), education-num (−0.46) và capital-gain (−0.36). 
Ba phương pháp LIME, DiCE và SHAP cùng hội tụ một kết luận: tuổi đóng vai trò như biến kích hoạt quyết định, và mô hình ưu tiên các yếu tố cấu trúc xã hội hơn các yếu tố phản ánh nỗ lực cá nhân.


 Tác động của sex_binary sau khi đã kiểm soát theo education-num. Toàn bộ cụm sex_binary = 0 (nữ) nằm trong vùng SHAP âm, kéo xác suất thu nhập cao giảm tới khoảng −0.5, kể cả ở nhóm có học vấn rất cao. Ngược lại, sex_binary = 1 (nam) trải dài sang vùng dương và có thể đạt khoảng +0.4. Điều này cho thấy học vấn giúp cải thiện dự đoán cho cả hai nhóm, nhưng không đủ để triệt tiêu hoàn toàn chênh lệch giới tính trong mô hình.

Ngay cả các điểm học vấn cao trong nhóm thiểu số cũng khó vượt sang vùng dương, cho thấy chênh lệch sắc tộc vẫn tồn tại khi các yếu tố năng lực đã được kiểm soát. Điều này chứng tỏ  mô hình đang học các mẫu phân bố thu nhập mang tính cấu trúc từ dữ liệu lịch sử, và các biến năng lực đơn lẻ không đủ để loại bỏ hoàn toàn khoảng cách này.

# 4.Kết luận và hướng phát triển của bài toán. 

### 4.1 Kết luận 

1. Ý nghĩa của việc khám phá mối quan hệ giữa đặc trưng và thu nhập
- Phân tích cho thấy mô hình không chỉ học năng lực cá nhân mà chủ yếu phản ánh cấu trúc phân bổ thu nhập trong xã hội. Các biến gắn với giai đoạn cuộc đời và vị thế xã hội có ảnh hưởng mạnh hơn các biến năng lực thuần túy. Trong ba yếu tố chính được quan tâm, age là biến quan trọng nhất, thể hiện tác động lớn và phi tuyến rõ rệt (dạng inverted-U). Marital status và occupation cũng có tác động mạnh, trong khi education có ảnh hưởng thấp hơn đáng kể. Điều này cho thấy bằng cấp đóng vai trò cần nhưng chưa đủ; kinh nghiệm, vị trí nghề nghiệp và giai đoạn sự nghiệp mới là yếu tố quyết định khả năng vượt ngưỡng thu nhập.

2. Tiền xử lý dữ liệu và mô hình mang lại hiệu suất tốt nhất
- Kết quả thực nghiệm cho thấy pipeline gồm feature engineering + xử lý mất cân bằng bằng SMOTE + LightGBM đạt hiệu suất và độ ổn định cao nhất. Feature engineering giúp mô hình học được các tương tác xã hội-kinh tế quan trọng; SMOTE cải thiện khả năng nhận diện lớp thu nhập cao; LightGBM tận dụng tốt quan hệ phi tuyến và tương tác giữa biến. Đồng thời, mô hình vẫn có thể giải thích được nhờ SHAP và LIME, giúp cân bằng giữa hiệu suất và khả năng diễn giải.

3. Mô hình có phản ánh định kiến lịch sử không?
- Phân tích fairness cho thấy tồn tại bias rõ rệt theo giới tính và sắc tộc. Tuy nhiên, SHAP cho thấy các biến nhạy cảm có tầm quan trọng trực tiếp thấp. Điều này chỉ ra rằng mô hình không sử dụng trực tiếp biến nhạy cảm, mà học định kiến thông qua proxy variables như marital status, occupation và capital features. Nhờ SHAP (giải thích toàn cục và cục bộ) và LIME (giải thích từng cá thể), có thể thấy quyết định của mô hình thường được thúc đẩy bởi các biến xã hội-nhân khẩu học hơn là các yếu tố năng lực.

4. Ảnh hưởng của giới tính và sắc tộc khi đã kiểm soát học vấn và giờ làm việc
- Ngay cả khi kiểm soát education và hours-per-week, chênh lệch theo giới tính và sắc tộc vẫn tồn tại. Điều này cho thấy sự khác biệt không chỉ đến từ nỗ lực cá nhân mà được nhúng trong cấu trúc dữ liệu xã hội-kinh tế mà mô hình học được. Nói cách khác, bất bình đẳng không nằm ở biến nhạy cảm riêng lẻ mà nằm trong toàn bộ mạng lưới đặc trưng liên quan đến vị thế xã hội.

### 4.2 hướng phát triển của bài toán. 

- Trong tương lai, hiệu suất mô hình có thể được cải thiện thông qua các kỹ thuật ensemble như weighted averaging hoặc stacking giữa LightGBM, XGBoost và CatBoost. Việc kết hợp nhiều mô hình giúp giảm phương sai dự đoán, tăng độ ổn định và khả năng tổng quát hóa. Bên cạnh đó, thay vì sử dụng một ngưỡng phân loại cố định, nghiên cứu nên phát triển cơ chế dynamic threshold cho các kịch bản vận hành khác nhau. Ngưỡng dự đoán có thể được điều chỉnh theo mục tiêu kinh doanh hoặc mức chấp nhận rủi ro, từ đó giúp hệ thống linh hoạt và phù hợp hơn với môi trường triển khai thực tế.

- Kết quả phân tích cho thấy chỉ một số đặc trưng đóng vai trò quan trọng trong dự đoán thu nhập. Do đó, hướng nghiên cứu tiếp theo nên tập trung vào việc lựa chọn tập đặc trưng tối thiểu nhưng vẫn đảm bảo hiệu năng. Các phương pháp như Recursive Feature Elimination kết hợp với SHAP có thể được sử dụng để xác định những biến quan trọng nhất. Việc giảm số lượng biến đầu vào giúp mô hình dễ giải thích, giảm chi phí tính toán và thuận lợi hơn trong quá trình bảo trì và triển khai.


- Giới hạn hiệu năng hiện tại cho thấy dữ liệu nhân khẩu học đơn thuần chưa đủ để dự đoán thu nhập chính xác. Trong tương lai, cần mở rộng tập dữ liệu bằng cách bổ sung các yếu tố theo thời gian như quá trình phát triển nghề nghiệp hoặc chu kỳ kinh tế. Ngoài ra, các nguồn dữ liệu bên ngoài như chỉ số kinh tế vùng, tốc độ tăng trưởng ngành và chất lượng giáo dục có thể cung cấp thêm tín hiệu dự đoán quan trọng. Việc tích hợp các nguồn dữ liệu này sẽ giúp mô hình phản ánh tốt hơn sự biến động thu nhập trong bối cảnh kinh tế – xã hội.


- SMOTE chỉ là một trong nhiều phương pháp xử lý mất cân bằng lớp. Nghiên cứu trong tương lai nên đánh giá các hướng tiếp cận khác như cost-sensitive learning hoặc focal loss, trong đó hàm mất mát được điều chỉnh để chú trọng hơn vào lớp thiểu số. Đồng thời, cần phân tích tác động của các phương pháp cân bằng dữ liệu đến độ ổn định của các kỹ thuật giải thích mô hình như SHAP và LIME. Điều này giúp cải thiện cả hiệu năng dự đoán lẫn độ tin cậy của các phân tích giải thích.
